```python
import numpy as np
import pandas as pd
```

```python
drc_data = pd.read_csv('drc-data-orig.csv')
```

```python
drc_data
```

```python
drc_data['AD.Date'] = pd.to_datetime(drc_data['AD.Date'])
drc_data['S2.DateSymptoms'] = pd.to_datetime(drc_data['S2.DateSymptoms'])
```

```python
np.mean(drc_data['S2.DateSymptoms'] - drc_data['AD.Date'])
```

```python
drc_data = drc_data[drc_data['LOS_bin'] != '<24h']
```

```python
bleeding = drc_data['S2.Bleeding']
any_bleeding = drc_data['S2.BleedingGums'] | drc_data['S2.BleedingInjection'] | drc_data['S2.BleedingNose'] | drc_data['S2.BloodStool'] | drc_data['S2.BloodVomit'] | drc_data['S2.BlackVomit'] |   drc_data['S2.BloodCough'] | drc_data['S2.NonMenstrualBlood'] | drc_data['S2.Hematoma'] | drc_data['S2.BloodUrine'] | drc_data['S2.BloodOther']
max(abs(bleeding - any_bleeding))
```

```python
drc_data['S2.JoinMusclePain'] = drc_data['S2.MusclePain'] | drc_data['S2.JointPain']
```

```python
drc_data['S7.FinalStatus'] = 1 - drc_data['S7.FinalStatus']
```

```python
drc_data.to_csv('drc-data.csv', index=False, na_rep="\\N")
```
